```python
import pandas as pd
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
from sklearn.decomposition import PCA

%matplotlib notebook
```

```python
file = "../cardio_dosis_data/cardio_clean_clustering_IMC.csv"
```

```python
data = pd.read_csv(file, 
                   usecols={'EDAD','PESO', 'TALLA', 'TIEMPO',
                            'mGy','pGym', 'IMÁGENES'})
```

```python
data.head()
```

```python
type(data)
data.isin(['NaN']).any()
```

```python
versus_cd_val = data.values
len(versus_cd_val)
```

```python
plt.scatter(versus_cd_val[:,4],versus_cd_val[:,2])
plt.title("CD vs TALLA")
```

```python
fig = sns.jointplot(versus_cd_val[:,4],versus_cd_val[:,0], 
              kind="kde", space=0, color="r", cbar=False)
fig.set_axis_labels('x', 'y', fontsize=16)
fig.ax_joint.set_xlabel('CD', fontweight='bold')
fig.ax_joint.set_ylabel('EDAD', fontweight='bold')
plt.suptitle("CD vs EDAD")
plt.savefig("CD1.png")
```

```python
fig = sns.jointplot(versus_cd_val[:,4],versus_cd_val[:,1], 
              kind="kde", space=0, color="g", cbar=False)
fig.set_axis_labels('x', 'y', fontsize=16)
fig.ax_joint.set_xlabel('CD', fontweight='bold')
fig.ax_joint.set_ylabel('PESO', fontweight='bold')
plt.suptitle("CD vs PESO")
plt.savefig("CD2.png")
```
